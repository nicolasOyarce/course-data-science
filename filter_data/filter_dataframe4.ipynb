4. Create a column with two conditionals or more options (np.create)

```python
import pandas as pd
import numpy as np
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin-1')
```

```python
# Create conditions and values
conditions = [
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    df_laptops['Price_euros']<=800,
]
#>3000,>2000,>800,<=800
values = ['Very expensive', 'Expensive', 'Cheap', 'Very Cheap']
```

```python
df_laptops['Price_levels'] = np.select(conditions, values, default='Not categorized')
```

```python
df_laptops
```

```python
df_laptops.value_counts('Price_levels')
```

```python
conditions_inches = [
    df_laptops['Inches']>16,
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    df_laptops['Inches']<=12,
]

values_inches = ['Very Big', 'Big', 'Small', 'Very Small']
```

```python
df_laptops['Inches_category'] = np.select(conditions_inches, values_inches, default='Not categorized')
```

```python
df_laptops.value_counts('Inches_category')
```
