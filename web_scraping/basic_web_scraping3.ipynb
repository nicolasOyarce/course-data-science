3. Organize Dictionary Data

```python
import pandas as pd
```

```python
root = "https://football-data.co.uk/mmz4281/"
```

```python
dict_countries = {
    'Spanish La Liga':'SP1',
    'Spanish Segunda Divicion':'SP2',
    'German Bundesliga':'D1',
    'English Premier League':'E0',
    'English League 1':'E2',
    'English League 2':'E3',
}
```

```python
dict_countries['Spanish La Liga']
```

```python
dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(18, 24):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv')
        df.insert(1, 'SEASON', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
