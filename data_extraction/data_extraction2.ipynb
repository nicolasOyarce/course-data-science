2. Method iloc()

```python
import pandas as pd
```

```python
df_players = pd.read_csv('players_20.csv')
```

```python
df_players.set_index('short_name', inplace=True)
df_players = df_players[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

Exercise 1

```python
df_players
```

```python
# height messi
df_players.iloc[0, 3]
```

```python
# weight cr7
df_players.iloc[1,4]
```

```python
# all columns height
df_players.iloc[:, 3]
```

```python
# all columns messi
df_players.iloc[0,:]
```

Exercise 2

```python
# data cr7 and messi
df_players.iloc[[0,1]]
```

```python
# height cr7 and messi
df_players.iloc[[0,1], 3]
```

```python
# height and weight messi
df_players.iloc[0,[3,4]]
```

```python
# height and weight messi and cr7
df_players.iloc[[0,1],[3,4]]
```

Exercise 3

```python
# slice
players = [0, 1]

df_players.iloc[players, 1:3]
```

```python
# slice with index name
columns = [1, 2, 3, 4]

df_players.index[:10]

df_players.iloc[0:10, columns]
```

Exercise 4

```python
# select players>180
columns = [1, 2, 3, 4]

df_players.iloc[list(df_players['height_cm']>180), columns]
```

```python
# players >180 and argentina
df_players[(df_players.iloc[:,3]>180) & (df_players.iloc[:, 5]=='Argentina')]
```
