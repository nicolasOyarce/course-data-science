9. Create document copy with copy()

```python
import pandas as pd
```

```python
df_players = pd.read_csv('players_20.csv')
```

```python
df_players.set_index('short_name', inplace=True)
df_players = df_players[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# deep=True
df_players_copy = df_players.copy()
```

```python
# update values in origins
df_players.loc['L. Messi', 'height_cm'] = 200
```

```python
# origin vs copy
df_players
df_players_copy
```

```python
# deep=False
df_players_shadow_copy = df_players.copy(deep=False)
```

```python
# update values in origins - shadow
df_players.loc['Cristiano Ronaldo', 'height_cm'] = 100
```

```python
df_players
df_players_shadow_copy
```

```python
# simple asignation = deep=False
df_new_copy = df_players
```

```python
df_players.loc['Neymar Jr', 'height_cm'] = 300
```

```python
df_new_copy
```
