3. Set values in cells

```python
import pandas as pd
import numpy as np
```

```python
df_players = pd.read_csv('players_20.csv')
```

```python
df_players.set_index('short_name', inplace=True)
df_players = df_players[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# set height messi
df_players.loc['L. Messi', 'height_cm'] = 175
```

```python
df_players
```

```python
# set value in all column
df_players.loc[:, 'height_cm'] = 120
```

```python
df_players
```

```python
# set update all row
df_players.iloc[-1, :] = np.nan
```

```python
df_players
```

```python
# set values multiple cells
df_players.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm']] = 190
```

```python
df_players.iloc[[0,1, 2, 3, 4], :]
```

```python
# set values for row meeting a condition
columns = ['age', 'dob', 'height_cm', 'weight_kg']

df_players.loc[df_players['height_cm'] < 180, columns] = 0
```

```python
df_players
```
