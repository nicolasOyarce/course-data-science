# Lambda

```python
import pandas as pd
```

```python
df_cars_sales = pd.read_csv('Car_sales.csv')
```

```python
df_cars_sales = df_cars_sales[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
df_cars_sales
```

```python
df_cars_sales.groupby('Manufacturer').sum().head(5)
```

```python
df_cars_sales.groupby('Manufacturer').sum().apply(lambda x:x*1000)[['Sales_in_thousands', 'Price_in_thousands']]
```

```python
# subtract values of the average in each
df_cars_sales.groupby('Manufacturer').apply(lambda x:x-x.mean(numeric_only=True))
```
