## Method .agg()

```python
df_car_sales.agg('sum')
```

```python
# calculate mean for numeric columns only
df_car_sales.select_dtypes(include='number').agg('mean')
```

```python
df_car_sales.select_dtypes(include='number').agg(['sum', 'mean'])
```

```python
df_car_sales.agg({
    'Sales_in_thousands':['sum', 'mean'],
    'Price_in_thousands':['sum', 'max']
})
```

```python
# sum in columns
df_car_sales[['Sales_in_thousands', 'Price_in_thousands']].agg('sum', axis=1)
```

```python
df_car_sales.agg(
    x=('Sales_in_thousands', 'sum'), 
    y=('Price_in_thousands', 'sum')
)
```
