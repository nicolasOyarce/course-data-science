# Groupby() and agg()

```python
import pandas as pd
```

```python
df_cars_sales = pd.read_csv('Car_sales.csv')
```

```python
df_cars_sales = df_cars_sales[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
df_cars_sales
```

```python
df_cars_sales.groupby('Vehicle_type').agg(['min', 'max'])
```

```python
df_cars_sales.groupby('Vehicle_type').agg( min_engine_size=('Engine_size', 'min'), max_horsepower=('Horsepower', 'max'))
```

```python
df_cars_sales.groupby('Manufacturer').agg( sum_sales = ('Sales_in_thousands', 'sum'), avg_price = ('Price_in_thousands', 'mean')).round(3)
```
