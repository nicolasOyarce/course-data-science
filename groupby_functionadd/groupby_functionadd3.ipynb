# Method GroupBy

```python
import pandas as pd
```

```python
df_cars_sales = pd.read_csv('Car_sales.csv')
```

```python
df_cars_sales = df_cars_sales[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
df_cars_sales
```

```python
# groupby replace split-apply-combine
df_cars_sales.groupby('Vehicle_type').mean(numeric_only=True)
```

```python
object_groupby = df_cars_sales.groupby('Manufacturer')
```

```python
object_groupby.groups.keys()
```

```python
object_groupby.get_group('Toyota')
```

```python
df_cars_sales.groupby('Manufacturer').mean(numeric_only=True)
```

```python
df_cars_sales.groupby('Vehicle_type').count()
```

```python
df_cars_sales.isnull().sum()
```

```python
df_cars_sales.groupby('Engine_size', dropna=False).count()
```
