# The strategy Split-Apply-Combine

```python
import pandas as pd
```

```python
df_cars_sales = pd.read_csv('Car_sales.csv')
```

```python
df_cars_sales = df_cars_sales[['Manufacturer', 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity']]
```

```python
df_cars_sales
```

```python
# SPLIT
car_filter = df_cars_sales['Vehicle_type']=='Car'
passenger_filter = df_cars_sales['Vehicle_type']=='Passenger'
```

```python
# APPLY
car_avg = df_cars_sales[car_filter]['Sales_in_thousands'].mean()
passenger_avg = df_cars_sales[passenger_filter]['Sales_in_thousands'].mean()
```

```python
# COMBINE
pd.DataFrame({
    'Vehicle_Type':['Car', 'Passenger'],
    'Sales_in_thousands':[car_avg, passenger_avg]
})
```
