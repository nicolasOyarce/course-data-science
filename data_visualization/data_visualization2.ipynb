```python
import pandas as pd 
import cufflinks as cf 
from IPython.display import display,HTML

cf.set_config_file(theme='pearl', sharing='public', offline=True)
```

```python
df_population = pd.read_csv('population_total.csv')
```

```python
# delete null values
df_population.dropna(inplace=True)

# delete duplicates
df_population.drop_duplicates(inplace=True)
```

```python
# create pivot table
df_population = df_population.pivot(index='year', columns='country', values='population')

# select contries
df_population = df_population[['United States', 'China', 'Indonesia', 'India', 'Brazil']]
```

```python
df_population
```

```python
# filter years out
df_population_sample = df_population[df_population.index.isin([1980, 1990, 2000, 2010, 2020])]

# plotting
df_population_sample.iplot(kind='bar', xTitle='Years', yTitle='Population')
```
