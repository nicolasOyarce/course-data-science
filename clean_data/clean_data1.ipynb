## Identify missing data

```python
# null values
df_netflix.isnull().sum().sort_values(ascending=False)
```

```python
# % missing values in each column
for column in df_netflix.columns:
    percentage = df_netflix[column].isnull().mean()
    print(column + ': ' + str(round(percentage*100, 2)) + '%')
```

## Manage missing data - .drop, .dropna, or .isnull

```python
# delete column
df_netflix.drop('director', axis=1)
```

```python
# delete row
not_director = df_netflix[df_netflix['director'].isnull()].index
df_netflix.drop(not_director, axis=0)
```

```python
# ~ + .isnull()
df_netflix[~(df_netflix['director'].isnull())]
```

```python
# dropna() - delete null values directly
df_netflix.dropna(subset=['director']).isnull().sum()
```

```python
# replace by average(numeric values) or moda(string values)
moda = ''.join(df_netflix['rating'].mode()) # 'TV-MA'
df_netflix['rating'].fillna(moda, inplace=True)
```

```python
df_netflix['rating'].isnull().sum()
```

```python
# replace by number aleatory
df_netflix['duration'].fillna(0, inplace=True)
df_netflix['duration'].isnull().sum() # 0 values
```

```python
# backward, forward
df_netflix.fillna(method='bfill') # rellena hacia atras
df_netflix.fillna(method='ffill') # rellena hacia adelante
```
